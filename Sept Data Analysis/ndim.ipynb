### Operation on array

```python
arr1 = np.array([45,67,12,40])
arr2 = np.array([10,20,30,40])
arr3 = arr1 + arr2
arr3 ## It will return the [+] operation of a both array
```

## Broadcasting

```python
arr1 = np.array([45,34,54,65])
arr2 = np.array([[23,34,67,87],
                [50,30,20,40]])
arr3 = arr1 + arr2
arr3 # it will execute the opeartion in both array [+] operator == [arr1 + arr2[i]] , [arr1 + arr2[ii]]
```

## Conditional selection

```python
arr1 = np.random.randint(1,100,15)
arr1
```

```python
arr1 < 50 ## It will return the array in a True & False
```

```python
arr1[arr1 < 50] ## it will return the array that less than 50.
```

```python
arr1 = np.random.randint(1,100,30).reshape(6,5)
arr1
```

```python
arr1[arr1 %2 ==0] ## It will return the even values in array.
```

```python
# Multiple of 3
arr1 [arr1 %3 ==0] ## It will return the number of multiplication of 3 number.
```

```python
arr1
```

```python
arr1 [(arr1 > 50 ) & (arr1 < 90)]
## In conditional selection we don't use [and] instead we use [&], and the 1 conditon is present then we don't have to put in paranthesis(),
# but if we have 2 condition then write in paranthesis().
```
